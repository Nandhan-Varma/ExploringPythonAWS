```python
import numpy as np 
import pandas as pd 
```

```python
dataFrame = pd.read_csv('/Dataset_Folder/MutualFunds.csv')
```

```python
dataFrame.head(5)
```

```python
#Finding out unique Fund Companies / Families
uniqueMutualFundFamily = pd.Series(dataFrame['fund_family']).unique() 
```

```python
print(uniqueMutualFundFamily)
```

```python
#Listing Out Various Parameters in The Dataset For Further Classification
print(list(dataFrame.columns))
```

```python
#Calculating Mean For Various Columns

meanOfColumns = dataFrame[['year_to_date_return','week52_high','week52_low','fund_return_1year','fund_return_3years','fund_return_5years']].mean()
```

```python
print(meanOfColumns)
```

```python
# Getting The Funds Which Belong to ACI
fundsOfACI = dataFrame[dataFrame['fund_family'] == 'American Century Investments']
```

```python
fundsOfACI.head(5) 
```

```python
print(fundsOfACI.shape) 
```

```python
fund_categories = pd.Series(fundsOfACI['fund_category']).unique()
```

```python
print(fund_categories)
```

```python
# Number of Funds Which Beats Mean Of One Year 

fundsOfACIWhichBeatsOneYearReturn = fundsOfACI[fundsOfACI['fund_return_1year'] > meanOfColumns['fund_return_1year']]

print(fundsOfACIWhichBeatsOneYearReturn.shape)
```

```python
# Name of the Funds Which Beats One Year Returns 
desiredColumnsData = ['fund_symbol','fund_short_name','week52_high','week52_low','fund_return_1year','fund_return_3years']
subsetOfMainDataFrame = fundsOfACIWhichBeatsOneYearReturn[desiredColumnsData]

requiredData = subsetOfMainDataFrame.head(10) 
print(requiredData)
```

```python
fundsOfACIWhichBeatsThreeYearReturn = fundsOfACI[fundsOfACI['fund_return_3years'] > meanOfColumns['fund_return_3years']]
print(fundsOfACIWhichBeatsThreeYearReturn.shape)
```

```python
fundsWhichBeatsBothOneYearandThreeYearReturns = fundsOfACIWhichBeatsOneYearReturn.merge(fundsOfACIWhichBeatsThreeYearReturn) 

print(fundsWhichBeatsBothOneYearandThreeYearReturns.shape)

```

```python
#Fund Names Which Beats Both One Year and Three Year Returns 
print(fundsOfACIWhichBeatsThreeYearReturn.head(10))
```

```python
fundsOfACIWhichBeatsFiveYearReturn = fundsOfACI['fund_return_5years'] > meanOfColumns['fund_return_5years']
```
